# Load Data

## Import Library

```python
import os
import glob
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)
```

## Menentukan lokasi dataset

```python
dataset_path = "../Datasheet/OBD-Anomaly-Detection/dataset"

print(dataset_path) 
```

## Melihat seluruh file csv

```python
csv_files = glob.glob(os.path.join(dataset_path, "*.csv"))

print("Jumlah File :", len(csv_files))

for file in csv_files:
    print(os.path.basename(file))
```

## Membaca seluruh dataset

```python
all_data = []

for file in csv_files:

    try:
        df = pd.read_csv(file)

        df["source_file"] = os.path.basename(file)

        all_data.append(df)

        print(os.path.basename(file),
              " --> ",
              df.shape)

    except Exception as e:

        print(file)
        print(e)
```

## Menggabungkan seluruh dataset

```python
dataset = pd.concat(all_data,
                    ignore_index=True)

print(dataset.shape)
```

## Menyimpan dataset gabungan

```python
dataset.to_csv("../Datasheet/OBD-Anomaly-Detection/dataset/all_dataset.csv",
               index=False)

print("Dataset berhasil disimpan.")
```
