# Preprocessing

## Import Library

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler

pd.set_option("display.max_columns", None)
```

## Load Dataset

```python
df = pd.read_csv("../Datasheet/OBD-Anomaly-Detection/dataset/all_dataset.csv")

print(df.shape)

df.head()
```

## Informasi Dataset

```python
df.info()
```

## Missing Value

```python
print(df.isnull().sum())
```

## Mengisi Missing Value

```python
numeric_columns = df.select_dtypes(include=np.number).columns

df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())

print(df.isnull().sum())
```

## Menghapus Data Duplikat

```python
print("Sebelum :", df.shape)

df = df.drop_duplicates()

print("Sesudah :", df.shape)
```

## Mengubah DateTime menjadi Format Waktu

```python
df["DateTime"] = pd.to_datetime(df["DateTime"])

df.head()
```

## Feature Selection

```python
features = [
    "lat",
    "lng",
    "MAF",
    "Speed",
    "GPH",
    "LPH",
    "KPL",
    "CO2 Emissions"
]

X = df[features]

X.head()
```

## Normalisasi Data

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

X_scaled = pd.DataFrame(X_scaled,
                        columns=features)

X_scaled.head()
```

## Membuat Label Anomaly

```python
df["Anomaly"] = (
    (df["Speed"] > 120) |
    (df["MAF"] > df["MAF"].quantile(0.99)) |
    (df["CO2 Emissions"] > df["CO2 Emissions"].quantile(0.99))
).astype(int)

df["Anomaly"].value_counts()
```

## Menggabungkan Data Hasil Scaling

```python
processed_data = X_scaled.copy()

processed_data["Anomaly"] = df["Anomaly"]

processed_data.head()
```

## Menyimpan Dataset

```python
processed_data.to_csv(
    "../Datasheet/OBD-Anomaly-Detection/dataset/preprocessing.csv",
    index=False
)

print("Preprocessing selesai.")
```

## Menampilkan Distribusi Label

```python
print(processed_data["Anomaly"].value_counts())

print("\nPersentase:")

print(processed_data["Anomaly"].value_counts(normalize=True) * 100)
```

Mengapa preprocessing ini penting?

* Membersihkan data dari nilai kosong dan duplikasi agar model tidak bias.
* Menyamakan skala fitur menggunakan StandardScaler sehingga algoritma berbasis jarak dan pohon bekerja lebih stabil.
* Membuat pseudo-label anomali sebagai dasar pelatihan model Random Forest karena dataset OBD-II tidak menyediakan label anomali secara langsung.
* Menghasilkan dataset yang konsisten dan siap digunakan pada tahap Isolation Forest dan Random Forest.
